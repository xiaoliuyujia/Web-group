```python
DFv = pd.read_csv("validation.csv")
#DF.head(10)
```

```python
def constModel(bid):
    bidcount = 0
    clicked = 0
    payp = []
        
    for index, row in DFv.iterrows():
        if bid > DFv.get_value(index, 'payprice'):
            bidcount+=1
            payp.append(DFv.get_value(index, 'payprice'))
            
            if DFv.get_value(index, 'click') > 0:
                clicked+=1
        
            
    return (bidcount, clicked, payp)    
```

```python
def RandModel():
    
    bidcount = 0
    clicked = 0
    payp = []
        
    for index, row in DFv.iterrows():
        bid = randint(227,300)
        if bid > DFv.get_value(index, 'payprice'):
            bidcount+=1
            payp.append(DFv.get_value(index, 'payprice'))
            
            if DFv.get_value(index, 'click') > 0:
                clicked+=1
        
            
    return (bidcount, clicked, payp)    
```

```python
nbids,numclick,paylist = constModel(273)    
```

```python
print('Number of bids: %s' % nbids)
print('Number of clicks: %s' % numclick)
print('Total cost: %s ' % sum(paylist))
```

```python
CTR(numclick, nbids)
```

```python
CPM(sum(paylist), nbids)
```

```python
CPC(sum(paylist), numclick)
```
